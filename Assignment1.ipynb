```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(start = -10, end = 10, steps = 10)
a = torch.rand(1)
b = torch.rand(1)
y = a * x + b
```

Y = aX + b

```python
plt.plot(x, y)
plt.scatter(x, y, color = 'red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("plot y = ax + b")
plt.grid()
```

Y = asin(x) + b

```python
x = torch.linspace(start = -10, end = 10, steps = 100)
a = torch.rand(1)
b = torch.rand(1)
y = a * torch.sin(x) + b
```

```python
plt.plot(x, y)
plt.scatter(x, y, color = 'red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('plot y = asin(x) + b')
```

Y = ae^X + b

```python
x = torch.linspace(start = -10, end = 10, steps = 100)
a = torch.rand(1)
b = torch.rand(1)
y = a * torch.exp(x) + b
```

```python
plt.plot(x, y)
plt.scatter(x, y, color = 'red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('plot y = a*e^x + b')
```

Y = aX^2 + b

```python
x = torch.linspace(start = -10, end = 10, steps = 100)
a = torch.rand(1)
b = torch.rand(1)
y = a * x ** 2 + b
```

```python
plt.plot(x, y)
plt.scatter(x, y, color = 'red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('plot y = ax^2 + b')
```
